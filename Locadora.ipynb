```python
import sqlite3
conn = sqlite3.connect('locadora.db')
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS cliente (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    nome TEXT,
    sobrenome TEXT,
    sexo TEXT,
    dataNascimento TEXT,
    cpf INTEGER,
    nomePai TEXT,
    nomeMae TEXT,
    endereco TEXT,
    complemento TEXT,
    bairro TEXT,
    cidade TEXT,
    telefone INTEGER,
    cep INTEGER
)
""")
conn.commit()
conn.close()

# CADASTRAR CLIENTE
def cadastrarCliente():
  conn = sqlite3.connect('locadora.db')
  cursor = conn.cursor()

  nome = input('Digite o nome do cliente: ')
  sobrenome = input('Digite o sobrenome do cliente: ')
  sexo = input('Digite o sexo do cliente: ')
  dataNascimento = input('Digite a data de nascimento do cliente: ')
  cpf = input('Digite o cpf do cliente: ')
  nomePai = input('Digite o nome do pai: ')
  nomeMae = input('Digite o nome da mãe: ')
  endereco = input('Digite o endereço: ')
  complemento = input('Digite o complemento: ')
  bairro = input('Digite o bairro: ')
  cidade = input('Digite a cidade: ')
  telefone = input('Digite o telefone: ')
  cep = input('Digite o cep: ')

  cursor.execute("""
  INSERT INTO cliente (nome, sobrenome, sexo, dataNascimento, cpf, nomePai, nomeMae, endereco, complemento, bairro, cidade, telefone, cep)
  VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
  """, (nome, sobrenome, sexo, dataNascimento, cpf, nomePai, nomeMae, endereco, complemento, bairro, cidade, telefone, cep))
  conn.commit()
  print('Cliente cadastrado com sucesso!')
  conn.close()

# LISTAR CLIENTES
def listarClientes():
  conn = sqlite3.connect('locadora.db')
  cursor = conn.cursor()
  cursor.execute('SELECT * FROM cliente')
  clientes = cursor.fetchall()
  for cliente in clientes:
    print(cliente)
  conn.close()

# ATUALIZAR CADASTRO
def atualizarCliente():
  conn = sqlite3.connect('locadora.db')
  cursor = conn.cursor()
  confirmacaoAtualizar = input("Gostaria de alterar os dados de algum cliente ? (s/n): ")
  if confirmacaoAtualizar.lower() == 's':
    idCliente = input("Digite o id do cliente que deseja atualizar: ")
    campo = input("Digite o campo que deseja atualizar: ")
    novoValor = input(f"Digite o novo valor para o campo {campo}: ")
    cursor.execute(f"UPDATE cliente SET {campo} = ? WHERE id = ?", (novoValor, idCliente))
    conn.commit()
    print("Cliente atualizado com sucesso!")


# DELETAR CADASTRO
def deletarCliente():
    conn = sqlite3.connect('locadora.db')
    cursor = conn.cursor()
    confirmacaoDeletar = input("Gostaria de deletar algum cliente ? (s/n): ")
    if confirmacaoDeletar.lower() == 'n':
        return
    idCliente = input("Digite o id do cliente que deseja deletar: ")

    confirmacaoDeletar = input(f"Tem certeza que deseja deletar o cliente com id {idCliente}? (s/n): ")
    if confirmacaoDeletar.lower() == 's':
        cursor.execute("DELETE FROM cliente WHERE id = ?", (idCliente,))
        conn.commit()
        print("Cliente deletado com sucesso!")
    conn.close()

def menu():
  while True:
    print("1. Cadastrar cliente")
    print("2. Listar clientes")
    print("3. Atualizar cliente")
    print("4. Deletar cliente")
    print("5. Sair")
    opcao = input("Escolha uma opção: ")
    if opcao == "1":
        cadastrarCliente()
    elif opcao == "2":
        listarClientes()
    elif opcao == "3":
        atualizarCliente()
    elif opcao == "4":
        deletarCliente()
    elif opcao == "5":
        break
    else:
        print("Opção inválida. Tente novamente.")
menu()
```
